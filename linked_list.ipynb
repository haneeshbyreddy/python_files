```python
class Node:
    def __init__(self,value):
        self.next=None
        self.value=value

class SingleLinkedList:
    def __init__(self) -> None:
        self.head=None
        self.count=0
    def add_at_beg(self,value):
        
        newnode=Node(value)
        
        if self.head is None:
            self.head=newnode

        else:
            newnode.next=self.head
            self.head=newnode
        self.count+=1
        
    def display(self):
        if self.head is None:
            print( "list is empty")
        else:
            current=self.head
            while(current):
                print(current.value)
                current=current.next
    def add_at_end(self,value):
        
        newnode=Node(value)
        if self.head is None:
            
            self.add_at_beg(value)
        else :
            
            current=self.head
            while(current.next is not None):
                current=current.next
            current.next=newnode
        self.count+=1
    def add_at_middle(self,value):
        newnode=Node(value)
        self.mid=self.count/2
        current=self.head
        while(self.mid-1):
            current=current.next
            self.mid-=1
        newnode.next=current.next
        current.next=newnode
        self.count+=1
        


node = SingleLinkedList()
node.add_at_beg(5)
node.display()
node.add_at_end(4)
node.display()
node.add_at_beg(6)
node.add_at_beg(15)
node.add_at_middle(7)
node.display()
```
