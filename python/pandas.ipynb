```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lambda_true = 1000  # True value of lambda
n_samples = 50  # Number of samples

# Generate data from an exponential distribution
np.random.seed(42)
data = np.random.exponential(scale=1/lambda_true, size=n_samples)

# Define the prior distribution (uniform)
lambda_values = np.linspace(800, 1200, 1000)  # Possible values for lambda
prior = np.full_like(lambda_values, 1/400)  # Uniform prior

# Calculate likelihood for each value of lambda
likelihood = np.exp(-lambda_values * data.sum()) * (lambda_values ** n_samples)

# Calculate the posterior (proportional to likelihood * prior)
posterior = likelihood * prior
posterior /= posterior.sum()  # Normalize to make the posterior a proper probability distribution

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(lambda_values, posterior, label='Posterior Distribution')
plt.axvline(lambda_true, color='red', linestyle='--', label='True Lambda')
plt.xlabel('Lambda')
plt.ylabel('Probability Density')
plt.title('Bayesian Inference with Uniform Prior')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Generate some synthetic data
np.random.seed(42)
lam_true = 2000
n_samples = 50
data = np.random.exponential(scale=1/lam_true, size=n_samples)

# Define the prior (uniform distribution)
lam_prior_min = 1  # Minimum value for lambda
lam_prior_max = 1000  # Maximum value for lambda

# Define the likelihood function (exponential distribution)
def likelihood(data, lam):
    return np.prod(expon.pdf(data, scale=1/lam))

# Define the posterior function
def posterior(data, lam_range, prior_range):
    lam_values = np.linspace(lam_range[0], lam_range[1], 1000)
    prior = 1 / (prior_range[1] - prior_range[0])  # Uniform prior
    likelihoods = [likelihood(data, lam) for lam in lam_values]
    unnormalized_posterior = prior * np.array(likelihoods)
    posterior = unnormalized_posterior / np.sum(unnormalized_posterior)
    return lam_values, posterior

# Calculate the posterior
lam_range = (lam_prior_min, lam_prior_max)
lam_values, posterior_values = posterior(data, lam_range, lam_range)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(lam_values, posterior_values, label='Posterior')
plt.axvline(x=lam_true, color='red', linestyle='--', label='True Lambda')
plt.xlabel('Lambda')
plt.ylabel('Posterior Probability Density')
plt.title('Bayesian Inference with Uniform Prior')
plt.legend()
plt.grid(True)
plt.show()
```
