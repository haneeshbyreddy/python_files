```python
import matplotlib.pyplot as plt
import numpy as np
import librosa
import sounddevice as sd
```

```python
def lms_noise_cancellation(primary_input, reference_input, filter_length, learning_rate, num_iterations):
    # Initialize filter coefficients with zeros
    w = np.zeros(filter_length)
    # Number of samples
    N = len(primary_input)
    
    # Initialize output signal
    output_signal = np.zeros(N)
    
    # Iterate through each sample
    for n in range(N):
        # Extract the current input samples for filtering
        x = np.zeros(filter_length)
        for i in range(filter_length):
            if n - i >= 0:
                x[i] = primary_input[n - i]
                                        
        # Calculate filter output
        y = np.dot(w, x)
      
        # Calculate error
        e = reference_input[n] - y
        
        # Update filter coefficients using LMS algorithm
        w = w + learning_rate * e * x
        
        # Store filtered output
        output_signal[n] = y
   


    
    return output_signal
```

```python
audio_file = r"E:\frnds\python\clean_speech\harvard.wav"

# Load the audio signal
audio_signal, sample_rate = librosa.load(audio_file, sr=None)

# Generate Gaussian noise with the same length as the audio signal
noise =0.1* np.random.randn(len(audio_signal))

# Add noise to the audio signal
noisy_input_signal = audio_signal + noise

# Parameters
filter_length = 200
learning_rate = 0.01
num_iterations = 100

# Apply LMS noise cancellation
output_signal = lms_noise_cancellation(noisy_input_signal, audio_signal, filter_length, learning_rate, num_iterations)

# Calculate SNR for LMS noise cancellation
original_error_lms = np.abs(audio_signal - output_signal)
noise_signal_lms = np.abs(noise)
snr_lms = 10 * np.log10(np.mean(np.square(audio_signal)) / np.mean(np.square(original_error_lms)))

# Calculate SNR for adaptive noise cancellation (assuming you have a function named adaptive_noise_cancelation)
#output_signal_adaptive = adaptive_noise_cancelation(noisy_input_signal, audio_signal, filter_length=100, mu=0.01, epsilon=0.01)
original_error_adaptive = np.abs(audio_signal - output_signal)
snr_adaptive = 10 * np.log10(np.mean(np.square(audio_signal)) / np.mean(np.square(original_error_adaptive)))

print("SNR for LMS Noise Cancellation:", snr_lms, "dB")
print("SNR for Adaptive Noise Cancellation:", snr_adaptive, "dB")
#output_signal=adaptive_noise_cancelation(noisy_input_signal, audio_signal, filter_length=100, mu=0.01, epsilon=0.01)
# Play original signal
print("Playing original signal...")
sd.play(audio_signal, sample_rate)
sd.wait()

# Play noisy input signal
print("Playing noisy input signal...")
sd.play(noisy_input_signal, sample_rate)
sd.wait()

# Play filtered output signal
print("Playing filtered output signal...")
sd.play(output_signal, sample_rate)
sd.wait()

# Plotting (if needed)
time = np.arange(len(audio_signal)) / sample_rate

plt.figure(figsize=(10, 6))
plt.grid(True)

plt.subplot(3, 1, 1)
plt.plot(time, noisy_input_signal, label='Noisy Input Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Noisy Input Signal')

plt.subplot(3, 1, 2)
plt.plot(time, audio_signal, label='Original Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Original Audio Signal')

plt.subplot(3, 1, 3)
plt.plot(time, output_signal, label='Filtered Output Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Filtered Audio Signal')

plt.tight_layout()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Gaussian distribution
mu = 0  # Mean
sigma = 1  # Standard deviation
num_samples = 1000  # Number of samples

# Generate random samples from a Gaussian distribution
samples = np.random.normal(mu, sigma, num_samples)

# Plotting the histogram of the samples
plt.figure(figsize=(8, 6))
plt.hist(samples, bins=30, density=True, alpha=0.6, color='b')

# Plot the probability density function (PDF) of the Gaussian distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
plt.plot(x, pdf, 'k', linewidth=2)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Gaussian Distribution')

plt.grid(True)
plt.show()
```

```python
import numpy as np
import soundfile as sf  # for reading/writing audio files
import sounddevice as sd  # for playing audio
def rls_algorithm(input_signal, desired_output, num_taps, forgetting_factor):
    num_samples = len(input_signal)
    w = np.zeros(num_taps)  # Initialize filter coefficients
    P = 1 / forgetting_factor * np.eye(num_taps)  # Initialize inverse covariance matrix

    estimated_output = np.zeros(num_samples)  # Initialize estimated output

    # Pad input signal to ensure consistent length of x_n
    padded_input_signal = np.concatenate((np.zeros(num_taps - 1), input_signal))

    for n in range(num_samples):
        x_n = padded_input_signal[n : n + num_taps]  # Extract input vector of length num_taps
        y_hat = np.dot(w, x_n)  # Predicted output

        error = desired_output[n] - y_hat  # Compute prediction error

        # Update filter coefficients
        P = (1 / forgetting_factor) * (P - np.outer(P @ x_n, x_n) @ P / (forgetting_factor + x_n @ P @ x_n))
        w = w + error * P @ x_n

        estimated_output[n] = y_hat  # Store estimated output

    return estimated_output


# Load audio file
audio_file = r"E:\frnds\python\clean_speech\harvard.wav"
input_signal, sample_rate = sf.read(audio_file)

# Convert to mono if needed (assuming stereo input)
if input_signal.ndim > 1:
    input_signal = np.mean(input_signal, axis=1)

# Example usage
num_taps = 10
forgetting_factor = 0.9

# Generate random desired output (for demonstration purposes)
true_coefficients = np.array([0.5, -0.3, 0.2, 0.1])
desired_output = np.convolve(input_signal, true_coefficients, mode='same')
desired_output += 0.1 * np.random.randn(len(desired_output))

# Apply RLS algorithm to estimate coefficients
estimated_output = rls_algorithm(input_signal, desired_output, num_taps, forgetting_factor)

# Generate output signal using estimated coefficients
output_signal = np.convolve(input_signal, estimated_output, mode='same')

# Play original signal
print("Playing original signal...")
sd.play(input_signal, sample_rate)
sd.wait()

# Play noisy input signal
print("Playing noisy input signal...")
sd.play(desired_output, sample_rate)
sd.wait()

# Play filtered output signal (generated from estimated coefficients)
print("Playing filtered output signal...")
sd.play(output_signal, sample_rate)
sd.wait()

# Print the true coefficients and estimated coefficients
print("True coefficients:", true_coefficients)
print("Estimated coefficients (RLS):", estimated_output[-num_taps:])
```
